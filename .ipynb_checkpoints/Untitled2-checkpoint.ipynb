```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('./data.csv')
dataset
```

## CLEANING UP THE DATA

```python
dataset.columns = dataset.columns.str.replace(' ','_')
dataset
```

```python
dataset.info()
```

```python
dataset.isnull()
```

```python
averageDATA = dataset.mean()
averageDATA
```

```python
dataset = dataset.fillna(averageDATA).drop_duplicates()
```

## MATCHING THE MEAN OF EACH COLUMN TO VARIABLES

```python
averageFA = dataset.fixed_acidity.mean()
averageVA = dataset.volatile_acidity.mean()
averageCA = dataset.citric_acid.mean()
averageRS = dataset.residual_sugar.mean()
averageC = dataset.chlorides.mean()
averageFSD = dataset.free_sulfur_dioxide.mean()
averageTSD = dataset.total_sulfur_dioxide.mean()
averageD = dataset.density.mean()
averagepH =dataset.pH.mean()
averageQ = dataset.quality.mean()
averageS = dataset.sulphates.mean()
averageA = dataset.alcohol.mean()
Indexing = dataset.index
fa = dataset.fixed_acidity
```

## MEAN OF EACH COLUMN

```python
dataset
```

```python
colors = ['red','orange','yellow','green','blue','purple','pink','brown','gray','black','violet','beige']
explode = (0.5, 0.5, 1, 1, 1, 0, 0, 0.5, 0, 0.5, 0, 0.5)
averageVALUES = [averageFA, averageVA, averageCA, averageRS, averageC, averageFSD, averageTSD, averageD, averagepH, averageS, averageA, averageQ]
averageTOPICS = ['FA',
                 'VA',
                 'CA',
                 'RS',
                 'C',
                 'FSD',
                 'TSD',
                 'density',
                 'pH',
                 'sulphates',
                 'alcohol',
                 'quality']
```

```python
xtitle = 'Columns names'
ytitle = 'Average values'
legendLabels = ['fixed_acidity(FA)',
                'volatile_acidity(VA)',
                'citric_acid(CA)',
                'residual_sugar(RS)',
                'chlorides(C)',
                'free_sulfur_dioxide(FSD)',
                'total_sulfur_dioxide(TSD)',
                'density',
                'pH',
                'sulphates',
                'alcohol',
                'quality']
```

```python
labels = averageTOPICS
sizes = averageVALUES
```

## COMPARING MEAN VALUES USING A PIE CHART

```python
plt.pie(sizes, labels = labels, explode = explode, autopct = '%1.1f%%', pctdistance = 1.01, shadow = True, colors = colors, labeldistance = 1.2)
plt.axis('equal')
plt.legend(title = 'index', labels = legendLabels, loc = 'upper left',)
plt.title('Average Values for each of the columns in a pie chart')
plt.figure(figsize = (15,18))
plt.show()
```

## COMPARING MEAN VALUES USING BAR GRAPH

```python
fig = plt.figure(figsize = (10,5))

plt.bar(labels, sizes, width = 0.5, color = colors)
plt.xlabel(xtitle)
plt.ylabel(ytitle)
plt.title('comparing mean of columns in a bar chart')
plt.legend(loc = 'upper right')
plt.show()
```
